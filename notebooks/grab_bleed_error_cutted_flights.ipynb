```python
def process_matching_files(dirpath, output_directory):
    # List to store the results
    matching_data_list = []
    
    # Loop through all files in the directory
    for filename in os.listdir(os.path.join(dirpath, 'Voos_com_erro')):
        if filename.startswith('voos_erro'):
            # Extract the numerical sequence from the filename
            sequence = filename.split('_')[2].split('.')[0]

            # Load the 'erro_teste' and 'cortado' files
            erro_file_path = os.path.join(os.path.join(dirpath, 'Voos_com_erro'), filename)
            cortado_file_path = os.path.join(os.path.join(dirpath, 'Voos_cortados'), f"voos_cortados_{sequence}.parquet")

            # Check if the corresponding 'cortado' file exists
            if os.path.exists(cortado_file_path):
                try:
                    erro_df = pd.read_parquet(erro_file_path)
                    cortado_df = pd.read_parquet(cortado_file_path)

                    # Find the matching values in the 'nome' column
                    matching_values = set(erro_df['nome']).intersection(set(cortado_df['nome']))

                    # Calculate the percentage of matching values
                    percentage = (len(matching_values) / len(erro_df)) * 100

                    # Store the results
                    matching_data_list.append({
                        "nome": filename,
                        "percentage": percentage
                    })

                except Exception as e:
                    print(f"Error processing the files {erro_file_path} and {cortado_file_path}: {e}")

    # Save the results to a .parquet file if there's any matching data
    if matching_data_list:
        matching_df = pd.DataFrame(matching_data_list)
        matching_file = os.path.join(output_directory, "matching_data.parquet")
        matching_df.to_parquet(matching_file)

    return matching_df if matching_data_list else None
```

```python
data_path = '/Users/henriquematias/Documents/GitHub/Inteli-Modulo-7/Projeto_Grupo1_Inteli_Azul/notebooks/Output'
output_path = '/Users/henriquematias/Documents/GitHub/Inteli-Modulo-7/Projeto_Grupo1_Inteli_Azul/notebooks/Output/Porcentagem_voos_erro_cortados'

process_matching_files(data_path, output_path)
```
